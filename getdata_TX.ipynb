```
from hikvisionapi import Client
from datetime import datetime
import cv2
import time
```

```
while True:
    now = datetime.now()
    dt_string = now.strftime("%d%m%Y_%H%M%S")
    print("date and time =", dt_string)
    cam = Client('http://182.52.58.44', 'admin', 'P@ssw0rd', timeout=10000)
    cam.count_events = 2 # The number of events we want to retrieve (default = 1)
    response = cam.Streaming.channels[102].picture(method='get', type='opaque_data')
    filenames = dt_string+'.jpg'
    print(filenames)
    with open('gdrive/My Drive/9_Tx/img_Tx/'+filenames, 'wb') as f:
        for chunk in response.iter_content(chunk_size=1024):
            if chunk:
                f.write(chunk)
    time.sleep(1000)
```
